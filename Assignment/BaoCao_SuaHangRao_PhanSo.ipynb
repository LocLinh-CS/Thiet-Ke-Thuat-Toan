2. **Pattern Recognition: tên dạng bài toán/kỹ thuật được áp dụng, các đặc điểm nhận dạng**
    - Tên dạng bài toán : vét can
    - Kĩ thuật áp dụng :

        + Dựa vào chiều cao đo lường được để tìm các hàng rào cần được đóng thêm ván.

    - Đặc điểm nhận dạng : các giá trị của hàng rào, tấm ván phải duyệt tuyến tính qua 1 lần duy nhất.
3. **Algorithm designed: thuật toán để giải quyết bài toán (trình bày bằng mã giả/ lưu đồ)**

4. **Đánh giá độ phức tạp thuật toán**
    - Time Complexity: O(nlog(n))

````python
def Find(a,brr,m,begin,high):
    while (begin!=m):
        if (a+brr[begin]>=high):
            break
        begin+=1
    if (begin==m):
        return -1
    return begin+1


n = int(input())
arr = [int(x) for x in input().split()]
m = int(input())
brr = [int(x) for x in input().split()]

a, b = sorted(arr), sorted(brr)
start, end = a[0]-1, a[-1]+b[-1]+1
if m >= n:
  start += b[m-n]
else:
  end = min(a[-1], min(a[0]+b[-1], a[m-1]+b[0]))+1
  
count_0=0
while (start!=end-1):
    high = int((start+end)/2)
    count=0
    begin=0
    check=1
    for index in range(n):
        if (arr[index]<high):
            if (begin==m):
                check=0
                break
            else:
                begin = Find(arr[index],brr,m,begin,high)
                if (begin==-1):
                    check=0
                    break
                else:
                    count+=1
    if (check==0):
        end=high
    else:
        count_0=count
        start=high
high=start
begin=0
print(high,count_0)

for index in range(n):
    if (arr[index]<high):
        begin = Find(arr[index],brr,m,begin,high)
        print(index+1, begin)
````

4. **Đánh giá độ phức tạp thuật toán**
    - Time Complexity: O(n)

````python
a = int(input())
b = int(input())
c = int(input())
d = int(input())

def gcd(tu, mau):
    while mau:
        tu, mau = mau, tu % mau
    return tu

def rutgon(tu, mau):
    uoc = gcd(tu, mau)
    if uoc != 1 :
        tu = tu / uoc
        mau = mau / uoc
    return tu , mau

def demsobuoc (tu1, mau1, tu2, mau2):
    dem = 0
    value1 = tu1 / mau1
    value2 = tu2 / mau2
    if (value1 > 1 and value2 < 1):
        print(0)
    elif(value1 < 1 and value2 > 1):
        print(0)
    else:
        while (tu1 != tu2 or mau1 != mau2 ):
            tu1 += 1
            mau1 += 1
            tu1 , mau1 = rutgon(tu1, mau1)
            dem += 1
            value_last = tu1 / mau1
            if value_last > 1 :
                if value_last < value2 :
                    dem = 0
                    break
            else:
                if value_last > value2 :
                    dem = 0
                    break
        print(dem)


demsobuoc(a, b, c, d)
````
